### Imports

```python
import pandas as pd
import numpy as np
import os
```

### Constants declared here

```python
#Constants

DATASET_PATH = os.path.normpath("D:\Data Science projects\Movies")
```

### The functions being used are defined in the cell below

```python
def load_dataset():
    data_set = {}
    data_set['movies_meta']  = pd.read_csv(os.path.join(DATASET_PATH, "movies_metadata.csv"), index_col=0)
    data_set['credits']  = pd.read_csv(os.path.join(DATASET_PATH, "credits.csv"), index_col=0)
    data_set['keywords'] = pd.read_csv(os.path.join(DATASET_PATH, "keywords.csv"), index_col=0)
    data_set['links'] = pd.read_csv(os.path.join(DATASET_PATH, "links.csv"), index_col=0)
    data_set['links_small'] = pd.read_csv(os.path.join(DATASET_PATH, "links_small.csv"), index_col=0)
    #data_set['ratings'] = pd.read_csv(os.path.join(DATASET_PATH, "ratings.csv"), index_col=0)
    data_set['ratings_small'] = pd.read_csv(os.path.join(DATASET_PATH, "ratings_small.csv"), index_col=0)

    return data_set
```

```python
data = load_dataset()
```
